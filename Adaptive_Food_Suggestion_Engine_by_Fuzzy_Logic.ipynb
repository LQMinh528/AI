```python
import numpy as np
import skfuzzy as fuzz
from skfuzzy import control as ctrl

Taste =ctrl.Antecedent(np.arange(0,11,1),'Taste')
Environment=ctrl.Antecedent(np.arange(0,11,1),'Environment')   
Budget=ctrl.Antecedent(np.arange(0,11,1),'Budget')
User_Review=ctrl.Antecedent(np.arange(0,11,1),'User_Review')
Location=ctrl.Antecedent(np.arange(0,11,1),'Location')
Level=ctrl.Consequent(np.arange(0,11,1),'Level') 

Taste['Low']=fuzz.trimf(Taste.universe,[0,0,5])
Taste['Medium']=fuzz.trimf(Taste.universe,[0,5,10])
Taste['High']=fuzz.trimf(Taste.universe,[5,10,10])

Environment['Low']=fuzz.trimf(Environment.universe,[0,0,5])
Environment['Medium']=fuzz.trimf(Environment.universe,[0,5,10])
Environment['High']=fuzz.trimf(Environment.universe,[5,10,10])

Budget['Less_Preferred']=fuzz.trimf(Budget.universe,[0,0,5])
Budget['Preferred']=fuzz.trimf(Budget.universe,[0,5,10])
Budget['More_Preferred']=fuzz.trimf(Budget.universe,[5,10,10])

User_Review['Less_Preferred']=fuzz.trimf(User_Review.universe,[0,0,5])
User_Review['Preferred']=fuzz.trimf(User_Review.universe,[0,5,10])
User_Review['More_Preferred']=fuzz.trimf(User_Review.universe,[5,10,10])

Location['Close']=fuzz.trimf(Location.universe,[0,0,5])
Location['Near']=fuzz.trimf(Location.universe,[0,5,10])
Location['Far']=fuzz.trimf(Location.universe,[5,10,10])

Level['Level 1']=fuzz.trimf(Level.universe,[0,2,4])
Level['Level 2']=fuzz.trimf(Level.universe,[2,4,6])
Level['Level 3']=fuzz.trimf(Level.universe,[4,6,8])
Level['Level 4']=fuzz.trimf(Level.universe,[6,8,10])
Level['Level 5']=fuzz.trimf(Level.universe,[8,10,10])

Taste.view()
Environment.view()
Budget.view()
User_Review.view()
Location.view()
Level.view()
```

```python
rule1=ctrl.Rule(Taste['Medium'] & Environment['Low'] & Budget ['Less_Preferred'] & Location ['Far'] & User_Review['Less_Preferred'],Level['Level 1'])
rule2=ctrl.Rule(Taste['Low'] & Environment['Medium'] & Budget ['Less_Preferred'] & Location ['Far'] & User_Review['Less_Preferred'],Level['Level 1'])
rule3=ctrl.Rule(Taste['Low'] & Environment['Low'] & Budget ['Preferred'] & Location ['Far'] & User_Review['Less_Preferred'],Level['Level 1'])
rule4=ctrl.Rule(Taste['Low'] & Environment['Low'] & Budget ['Less_Preferred'] & Location ['Near'] & User_Review['Less_Preferred'],Level['Level 1'])
rule5=ctrl.Rule(Taste['Low'] & Environment['Low'] & Budget ['Less_Preferred'] & Location ['Far'] & User_Review['Preferred'],Level['Level 1'])

rule6=ctrl.Rule(Taste['Low'] & Environment['Medium'] & Budget ['Preferred'] & Location ['Near'] & User_Review['Preferred'],Level['Level 2'])
rule7=ctrl.Rule(Taste['Medium'] & Environment['Low'] & Budget ['Preferred'] & Location ['Near'] & User_Review['Preferred'],Level['Level 2'])
rule8=ctrl.Rule(Taste['Medium'] & Environment['Medium'] & Budget ['Less_Preferred'] & Location ['Near'] & User_Review['Preferred'],Level['Level 2'])
rule9=ctrl.Rule(Taste['Medium'] & Environment['Medium'] & Budget ['Preferred'] & Location ['Far'] & User_Review['Preferred'],Level['Level 2'])
rule10=ctrl.Rule(Taste['Medium'] & Environment['Medium'] & Budget ['Preferred'] & Location ['Near'] & User_Review['Less_Preferred'],Level['Level 2'])

rule11=ctrl.Rule(Taste['Low'] & Environment['Low'] & Budget ['Preferred'] & Location ['Near'] & User_Review['Preferred'],Level['Level 2'])
rule12=ctrl.Rule(Taste['Medium'] & Environment['Low'] & Budget ['Less_Preferred'] & Location ['Near'] & User_Review['Preferred'],Level['Level 2'])
rule13=ctrl.Rule(Taste['Medium'] & Environment['Medium'] & Budget ['Less_Preferred'] & Location ['Far'] & User_Review['Preferred'],Level['Level 2'])
rule14=ctrl.Rule(Taste['Medium'] & Environment['Medium'] & Budget ['Preferred'] & Location ['Far'] & User_Review['Less_Preferred'],Level['Level 2'])
rule15=ctrl.Rule(Taste['Low'] & Environment['Medium'] & Budget ['Preferred'] & Location ['Near'] & User_Review['Less_Preferred'],Level['Level 2'])

rule16=ctrl.Rule(Taste['High'] & Environment ['Low'] & Budget ['Preferred'] & User_Review['Preferred'] & Location ['Near'] , Level['Level 3'])
rule17=ctrl.Rule(Taste['Medium'] & Environment ['High'] & Budget ['Less_Preferred'] & User_Review['Preferred'] & Location ['Near'] , Level['Level 3'])
rule18=ctrl.Rule(Taste['Medium'] & Environment ['Medium'] & Budget ['More_Preferred'] & User_Review['Less_Preferred'] & Location ['Near'] , Level['Level 3'])
rule19=ctrl.Rule(Taste['Medium'] & Environment ['Medium'] & Budget ['Preferred'] & User_Review['More_Preferred'] & Location ['Far'] , Level['Level 3'])
rule20=ctrl.Rule(Taste['Low'] & Environment ['Medium'] & Budget ['Preferred'] & User_Review['Preferred'] & Location ['Close'] , Level['Level 3'])

rule21=ctrl.Rule(Taste['High'] & Environment['Medium'] & Budget ['Preferred'] & Location ['Near'] & User_Review['Preferred'],Level['Level 4'])
rule22=ctrl.Rule(Taste['Medium'] & Environment['High'] & Budget ['Preferred'] & Location ['Near'] & User_Review['Preferred'],Level['Level 4'])
rule23=ctrl.Rule(Taste['Medium'] & Environment['Medium'] & Budget ['More_Preferred'] & Location ['Near'] & User_Review['Preferred'],Level['Level 4'])
rule24=ctrl.Rule(Taste['Medium'] & Environment['Medium'] & Budget ['Preferred'] & Location ['Close'] & User_Review['Preferred'],Level['Level 4'])
rule25=ctrl.Rule(Taste['Medium'] & Environment['Medium'] & Budget ['Preferred'] & Location ['Near'] & User_Review['More_Preferred'],Level['Level 4'])

rule26=ctrl.Rule(Taste['High'] & Environment['High'] & Budget ['Preferred'] & Location ['Near'] & User_Review['Preferred'],Level['Level 4'])
rule27=ctrl.Rule(Taste['Medium'] & Environment['High'] & Budget ['More_Preferred'] & Location ['Near'] & User_Review['Preferred'],Level['Level 4'])
rule28=ctrl.Rule(Taste['Medium'] & Environment['Medium'] & Budget ['More_Preferred'] & Location ['Close'] & User_Review['Preferred'],Level['Level 4'])
rule29=ctrl.Rule(Taste['Medium'] & Environment['Medium'] & Budget ['Preferred'] & Location ['Close'] & User_Review['More_Preferred'],Level['Level 4'])
rule30=ctrl.Rule(Taste['High'] & Environment['Medium'] & Budget ['Preferred'] & Location ['Near'] & User_Review['More_Preferred'],Level['Level 4'])

rule31=ctrl.Rule(Taste['Medium'] & Environment['High'] & Budget ['More_Preferred'] & Location ['Close'] & User_Review['More_Preferred'],Level['Level 5'])
rule32=ctrl.Rule(Taste['High'] & Environment['Medium'] & Budget ['More_Preferred'] & Location ['Close'] & User_Review['More_Preferred'],Level['Level 5'])
rule33=ctrl.Rule(Taste['High'] & Environment['High'] & Budget ['Preferred'] & Location ['Close'] & User_Review['More_Preferred'],Level['Level 5'])
rule34=ctrl.Rule(Taste['High'] & Environment['High'] & Budget ['More_Preferred'] & Location ['Near'] & User_Review['More_Preferred'],Level['Level 5'])
rule35=ctrl.Rule(Taste['High'] & Environment['High'] & Budget ['More_Preferred'] & Location ['Close'] & User_Review['Preferred'],Level['Level 5'])
rule36=ctrl.Rule(Taste['Medium'] & Environment ['Medium'] & Budget ['Preferred'] & User_Review['Preferred'] & Location ['Near'] , Level['Level 3'])
```

```python
leveling_ctrl=ctrl.ControlSystem([rule1,rule2,rule3,rule4,rule5,rule6,rule7,rule8,rule9,rule10,rule11,rule12,rule13,rule14,rule15,rule16,rule17,rule18,rule19,rule20,rule21,rule22,rule23,rule24,rule25,rule26,rule27,rule28,rule29,rule30,rule31,rule32,rule33,rule34,rule35,rule36])
leveling=ctrl.ControlSystemSimulation(leveling_ctrl)
leveling.input['Taste']= 3
leveling.input['Environment']=2
leveling.input['Budget']=4             
leveling.input['User_Review']=8
leveling.input['Location']= 6
leveling.compute()
Level.view(sim = leveling)
```
