```python
from sklearn import datasets                             #Thư viện lấy dữ liệu
from sklearn.linear_model import Perceptron              #Lấy model Perceptron
from sklearn.model_selection import train_test_split     #Hàm train test split
from sklearn.metrics import accuracy_score               
from sklearn.preprocessing import StandardScaler

```

```python
iris = datasets.load_iris()
X = iris.data
y = iris.target
print(X[:5])
print(y[:5])
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size = 0.3)
```

```python
sc = StandardScaler()
sc.fit(X_train)
X_train_std = sc.transform(X_train)
X_test_std = sc.transform(X_test)
```

```python
print(X_train, X_train_std)
```

```python
#max_iter = số lần học
#eta0 = tốc độ học
#random_state = lấy ngẫu nhiên
model = Perceptron(max_iter = 1000, eta0 = 0.001, random_state = 0)
model.fit(X_train, y_train)
y_predict = model.predict(X_test)
print(y_predict)
print(y_test)
print('Độ chính xác mô hình chưa scale: ', accuracy_score(y_test, y_predict))

model = Perceptron(max_iter = 1000, eta0 = 0.001, random_state = 0)
model.fit(X_train_std, y_train)
y_predict = model.predict(X_test_std)
print(y_predict)
print(y_test)
print('Độ chính xác mô hình: ', accuracy_score(y_test,y_predict))
```
